```python
def flattenImage(label,name):
    x_data = np.array( [np.array(cv.imread('TEST\\'+ label+"\\"+name)) for i in range(len('TEST\\'+ label +"\\"+name))] )
    pixels = x_data.flatten()
    return pixels
```
